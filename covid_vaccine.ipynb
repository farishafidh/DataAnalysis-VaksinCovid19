```python
q1 = dataset.copy()
q1.drop(['daily_vaccinations_raw', 'daily_vaccinations', 'total_vaccinations_per_hundred', 
         'people_vaccinated_per_hundred',
         'people_fully_vaccinated_per_hundred', 'daily_vaccinations_per_million',
         'source_name', 'source_website'], axis=1, inplace=True)
```

### Menambahkan nama negara ke dalam data 'List Vaccines'

```python
list_countries_df = pd.read_excel(pwd + '\\modfied_country_vaccinations.xlsx',
                                  sheet_name='Latest Vaccination')
```

```python
list_vaccines_df = pd.DataFrame(columns=['vaccine','country','number_of_countries'])
vaccines_list = list_countries_df['vaccines'].str.split(', ')
unique_vaccines = list(set(vaccine for sublist in vaccines_list for vaccine in sublist))

for vaccine in unique_vaccines:
    countries = list(list_countries_df[list_countries_df['vaccines'].str.contains(vaccine)]['country'])
    num_countries = list_countries_df[list_countries_df['vaccines'].str.contains(vaccine)]['country'].nunique()
    
    for country in countries:
        list_vaccines_df = list_vaccines_df.append({'vaccine':vaccine, 'country':country, 'number_of_countries':num_countries},
                                 ignore_index=True)
    
```

```python
list_vaccines_df.sort_values(by='vaccine', ascending=True, inplace=True)
list_vaccines_df.reset_index(drop=True, inplace=True)

list_vaccines_df
```

```python
with pd.ExcelWriter(pwd + '\\modfied_country_vaccinations.xlsx', engine='openpyxl', mode='a') as writer:
    list_vaccines_df.to_excel(writer, sheet_name='List Vaccines New', index=False)
```
